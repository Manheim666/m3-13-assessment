```python
def format_p(p):
    return "<0.001" if p < 0.001 else round(p, 4)

def corr_report(df, pairs):
    out = []
    for a, b in pairs:
        pear_r, pear_p = stats.pearsonr(df[a], df[b])
        spear_r, spear_p = stats.spearmanr(df[a], df[b])

        out.append({
            "pair": f"{a} vs {b}",
            "pearson_r": round(pear_r, 4),
            "pearson_p": format_p(pear_p),
            "spearman_r": round(spear_r, 4),
            "spearman_p": format_p(spear_p)
        })

    return pd.DataFrame(out)


pairs = [
    ("mpg", "horsepower"),
    ("mpg", "weight"),
    ("weight", "acceleration"),
    ("horsepower", "acceleration"),
    ("weight", "horsepower")
]

corr_df = corr_report(mpg, pairs)
corr_df
```

## Task 2 — Group Comparisons

1. Choose a numeric variable and visualize its distribution by origin (box/violin plot).
2. Check ANOVA assumptions: Shapiro-Wilk (normality) and Levene's test (equal variances).
3. Run one-way ANOVA — report F-statistic, p-value, and eta-squared.
4. If significant, run Tukey's HSD and state which groups differ.

```python
# Distribution visualization by origin
plt.figure(figsize=(8,5))
sns.boxplot(data=mpg, x="origin", y="mpg")
plt.title("MPG distribution by origin (Boxplot)")
plt.show()

plt.figure(figsize=(8,5))
sns.violinplot(data=mpg, x="origin", y="mpg")
plt.title("MPG distribution by origin (Violin plot)")
plt.show()
```

```python
# Assumption checks: Shapiro-Wilk + Levene's test
usa = mpg[mpg["origin"] == "usa"]["mpg"]
europe = mpg[mpg["origin"] == "europe"]["mpg"]
japan = mpg[mpg["origin"] == "japan"]["mpg"]

print("\nShapiro-Wilk Test (Normality)")
for origin in mpg["origin"].unique():
    data = mpg[mpg["origin"] == origin]["mpg"]
    stat, p = stats.shapiro(data)
    print(f"{origin:8s} → W={stat:.4f}, p={p:.4f}")

print("\nLevene Test (Equal Variances)")
stat, p = stats.levene(usa, europe, japan)
print(f"stat={stat:.4f}, p={p:.4f}")
```

```python
# One-way ANOVA
stat, p = stats.f_oneway(usa, europe, japan)

print(f"ANOVA F-statistic: {stat:.4f}")
print(f"ANOVA p-value: {p:.6f}")


grand_mean = mpg["mpg"].mean()
groups = [usa, europe, japan]
ss_between = sum(len(g) * (g.mean() - grand_mean)**2 for g in groups)
ss_total = ((mpg["mpg"] - grand_mean) ** 2).sum()

eta_squared = ss_between / ss_total
print("Eta-squared:", eta_squared)
```

```python
# Post-hoc test (Tukey's HSD)
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=mpg["mpg"], groups=mpg["origin"], alpha=0.05)
print(tukey)
```

## Task 3 — Classification with Bootstrap Confidence Intervals

1. Prepare features (numeric columns) and target (`origin`). Train/test split (75/25, stratified).
2. Train at least two classifiers with default hyperparameters.
3. Print a classification report for each.
4. Write `bootstrap_metric()` and compute 95 % CIs for F1, precision, and recall.
5. Visualize the CIs and state which model you recommend.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score

numeric_cols = ["mpg", "cylinders", "displacement", "horsepower",
                "weight", "acceleration", "model_year"]

X = mpg[numeric_cols].values
y = mpg["origin"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=SEED
)

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
print(f"Class distribution (test): {pd.Series(y_test).value_counts().to_dict()}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# LOGREG
log_reg = LogisticRegression(max_iter=5000, solver="lbfgs")
log_reg.fit(X_train_scaled, y_train)
log_pred = log_reg.predict(X_test_scaled)

#RANDOM FOREST
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

# KNN
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
knn_pred = knn.predict(X_test_scaled)

```

```python
# Classification reports
from sklearn.metrics import classification_report

print("LOGISTIC REGRESSION")
print(classification_report(y_test, log_pred))

print("\nRANDOM FOREST")
print(classification_report(y_test, rf_pred))

print("\nK-NEAREST NEIGHBORS")
print(classification_report(y_test, knn_pred))
```

```python
def bootstrap_metric(y_true, y_pred, metric_fn, n_boot=2000, seed=42):
    """
    Compute bootstrap 95% confidence interval for a classification metric.

    Returns:
    ----------
    point_estimate : float
    ci_lower : float
    ci_upper : float
    boot_scores : np.ndarray
    """

    np.random.seed(seed)

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    n = len(y_true)
    scores = []

    for _ in range(n_boot):
        idx = np.random.choice(n, n, replace=True)

        yt = y_true[idx]
        yp = y_pred[idx]

        score = metric_fn(yt, yp, average="weighted", zero_division=0)
        scores.append(score)

    scores = np.array(scores)

    point_estimate = metric_fn(y_true, y_pred, average="weighted", zero_division=0)
    ci_lower = np.percentile(scores, 2.5)
    ci_upper = np.percentile(scores, 97.5)

    return point_estimate, ci_lower, ci_upper, scores
```

```python
# Apply bootstrap_metric for F1, precision, and recall (each model)
from sklearn.metrics import f1_score, precision_score, recall_score

metrics = {
    "F1": f1_score,
    "Precision": precision_score,
    "Recall": recall_score
}

print("LOGISTIC REGRESSION")

for name, fn in metrics.items():
    mean, low, high, _ = bootstrap_metric(y_test, log_pred, fn)
    print(f"{name}: {mean:.3f} (95% CI: [{low:.3f} : {high:.3f}])")

print("\nRANDOM FOREST")

for name, fn in metrics.items():
    mean, low, high, _ = bootstrap_metric(y_test, rf_pred, fn)
    print(f"{name}: {mean:.3f} (95% CI: [{low:.3f} : {high:.3f}])")

print("\nK-NEAREST NEIGHBORS")
for name, fn in metrics.items():
    mean, low, high, _ = bootstrap_metric(y_test, knn_pred, fn)
    print(f"{name}: {mean:.3f} (95% CI: [{low:.3f} : {high:.3f}])")
```

```python
# Visualize bootstrap CIs (dot-and-whisker or grouped bar chart)

def get_bootstrap_results(y_true, y_pred):
    results = {}
    
    for name, fn in metrics.items():
        mean, low, high, _ = bootstrap_metric(y_true, y_pred, fn)
        results[name] = (mean, low, high)

    return results


# Compute bootstrap results
lr = get_bootstrap_results(y_test, log_pred)
rf = get_bootstrap_results(y_test, rf_pred)
knn = get_bootstrap_results(y_test, knn_pred)

# Plotting
metrics_list = ["F1", "Precision", "Recall"]

fig, ax = plt.subplots(figsize=(10, 5))

x = np.arange(len(metrics_list))
offset = 0.15


def plot_model(data, offset, label):
    means = [data[m][0] for m in metrics_list]
    lows = [data[m][1] for m in metrics_list]
    highs = [data[m][2] for m in metrics_list]

    yerr = [
        np.array(means) - np.array(lows),
        np.array(highs) - np.array(means)
    ]

    ax.errorbar(
        x + offset,
        means,
        yerr=yerr,
        fmt='o',
        capsize=5,
        label=label
    )


plot_model(lr, -offset, "Logistic Regression")
plot_model(rf, offset, "Random Forest")
plot_model(knn, 3 * offset, "K-Nearest Neighbors")
ax.set_xticks(x)
ax.set_xticklabels(metrics_list)
ax.set_ylabel("Score")
ax.set_title("Bootstrap Confidence Intervals (Corrected)")
ax.legend()

plt.tight_layout()
plt.show()
```
